```python
# First things first, import the urllib3 module:

import urllib.request
site = "https://www.imdb.com/title/tt9376612/reviews?ref_=tt_urv"
#http://www.imdb.com/{data-ajaxurl}?paginationKey={data-key}
#site="http://www.imdb.com/title/tt9376612/reviews/_ajax?paginationKey=xxdr3yxoixpwgjg7c6vq44huu6hi6xe5qywg3adr43xlnl5sas6df5ed53iunemqakizscc53xq32"
#site = "http://www.imdb.com/{/title/tt9376612/reviews/_ajax}?paginationKey={g4wp7dzeqeytk3qc66vhfnzwrhr4qcjyy4hhzo5ziwr26fbyhvrl4ty4o4zf3mzidnrndtthpyuzlpjmgsvjsxsndjrrrk7r}"
page = urllib.request.urlopen(site)
from bs4 import BeautifulSoup
soup = BeautifulSoup(page)
reviews= soup.find_all(class_="lister-item-content")
    #print(reviews)
titles=[]
dates=[]
text=[]
rating=[] 
for a in soup.findAll('a', class_='title'):
            titles.append(a.text)
        
for a in soup.findAll(class_='display-name-date'):
            dates.append(a.text)
for a in soup.findAll(class_='content'):
            text.append(a.text)
for a in soup.findAll(class_='ipl-ratings-bar'):
            rating.append(a.text)
titles_clean = []
dates_clean=[]
text_clean=[]
rating_clean=[]
for element in titles:
    titles_clean.append(element.strip())
for element in dates:
    dates_clean.append(element.strip())
for element in text:
    text_clean.append(element.strip())
for element in rating:
    rating_clean.append(element.strip())
import pandas as pd
reviews = pd.DataFrame({
        "titles": titles_clean,
        "user name and dates":dates_clean,
        "ratings":rating_clean,
        "text":text_clean
})
reviews
```

Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/). 

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# You code here (Please add comments in the code):


import urllib.request
import requests
from bs4 import BeautifulSoup
import pandas as pd
heading = []
heading_description =[] 
b=[]
c=[]
article = input("Enter keyword to search: ") 
for number in range(2):
  access = requests.get("https://scholar.google.com/scholar?start="+str(number)+"&q="+article+"&hl=en&as_sdt=0,5")
  soup = BeautifulSoup(access.text, 'html.parser')
  review_username = soup.find_all('div',class_='gs_a') 
  review_title = soup.find_all('h3',class_='gs_rt')
  review_text = soup.find_all('div',class_='gs_rs')
  for element, sub_element,i in zip(review_username,review_title,review_text) : 
   heading.append(element.text) 
   heading_description.append(sub_element.text)
   b.append(i.text)
review_dataframe= pd.DataFrame(list(zip(heading,heading_description,b)), columns =['Heading', 'Heading Description','b'])
review_dataframe



  
```
